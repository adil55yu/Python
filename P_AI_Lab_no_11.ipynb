**Task 1:** Use the dataset given below of grades of students in a class. Calculate the frequency of each
grade and then create a histogram and a bar chart to visualize distribution of grades among the
students.      
Dataset: A, B, C, B, A, C, C, B, A, A, B.

```python
import matplotlib.pyplot as plt
from collections import Counter

grades = ['A', 'B', 'C', 'B', 'A', 'C', 'C', 'B', 'A', 'A', 'B']

frequency = Counter(grades)

print("Frequency of each grade:")
for grade, count in frequency.items():
    print(grade, ":", count)

plt.figure()
plt.bar(frequency.keys(), frequency.values())
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.title("Bar Chart of Grades")
plt.show()
```

```python
plt.figure()
plt.hist(grades)
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.title("Histogram of Grades")
plt.show()
```

**Task 2:** Find the frequency of alphabets from the text below and visualize it using a bar chart.      
Text: Airports are always a bustling entity with many emotions all over the place. Some are happy to
welcome their loved ones in their arms, some are sad to see them go, while others are anxiously waiting
for their flights to reach. Rolling suitcases, hurrying footsteps, laughing children and the electronic ding
of the machines, all accumulated in one place.

```python
import matplotlib.pyplot as plt
from collections import Counter
import string

text = """Airports are always a bustling entity with many emotions all over the place.
Some are happy to welcome their loved ones in their arms, some are sad to see them go,
while others are anxiously waiting for their flights to reach. Rolling suitcases,
hurrying footsteps, laughing children and the electronic ding of the machines,
all accumulated in one place."""

text = text.lower()
alphabets_only = [char for char in text if char in string.ascii_lowercase]

frequency = Counter(alphabets_only)

letters = sorted(frequency.keys())
counts = [frequency[letter] for letter in letters]

plt.figure()
plt.bar(letters, counts)
plt.xlabel("Alphabets")
plt.ylabel("Frequency")
plt.title("Frequency of Alphabets in Text")
plt.show()
```

**Task 3:** Find the mean, median, mode, variance, range, variance, standard deviation, quartiles (Q1, Q2,
Q3, IQR) and box plot of the data given below.      
Data: 23, 45, 56, 34, 78, 90, 12, 67, 43, 55, 88, 32, 67, 54, 78, 89, 22, 35, 66, 43.      
(Use the library of numpy, statistics and matplotlib for this question.      
Formula for quartiles: np.percentile (data, 25/50/75).
Formula for IQR: q3-q1 Formula for boxplot:
plt.boxplot(data)
Plt.show()

```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt

data = [23, 45, 56, 34, 78, 90, 12, 67, 43, 55, 88, 32, 67, 54, 78, 89, 22, 35, 66, 43]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
variance = np.var(data)
std_dev = np.std(data)
data_range = max(data) - min(data)

q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Range:", data_range)
print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)
print("IQR:", iqr)

plt.boxplot(data)
plt.show()
```

**Task 4:** Use the dataset of Laptops Price Dataset and solve the following questions.    
1. Print the header, information and describe the data.  
2. Find the number of brands in the dataset, display their counts and visualize them in a bar chart.   
3. Find the mean, median, mode and variance of the price of laptops and visualize it in a box plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics as stats

df = pd.read_csv("laptops.csv")

print(df.head())
print(df.info())
print(df.describe())

brand_counts = df["Brand"].value_counts()
print(brand_counts)

plt.figure()
plt.bar(brand_counts.index, brand_counts.values)
plt.xticks(rotation=90)
plt.xlabel("Brand")
plt.ylabel("Count")
plt.title("Number of Laptops by Brand")
plt.show()

prices = df["Final Price"]

mean_price = np.mean(prices)
median_price = np.median(prices)
mode_price = stats.mode(prices)
variance_price = np.var(prices)

print("Mean Price:", mean_price)
print("Median Price:", median_price)
print("Mode Price:", mode_price)
print("Variance:", variance_price)

plt.boxplot(prices)
plt.xlabel("Laptop Prices")
plt.title("Box Plot of Laptop Prices")
plt.show()
```
