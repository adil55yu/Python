4. Create a 2D array of shape (5, 5) with random integers between 0 and 9, and replace all even
values with -1.

```python
arr2 = np.random.randint(0, 10, (5, 5))
arr2[arr2 % 2 == 0] = -1
print(arr2)
```

5. Given two arrays a and b, find the common elements between them.

```python
a = np.array([1, 2, 3, 4, 5, 6])
b = np.array([4, 5, 6, 7, 8])

common = np.intersect1d(a, b)
print(common)
```

6. Calculate the matrix product of a 3x4 matrix and a 4x2 matrix.

```python
A = np.random.randint(1, 10, (3, 4))
B = np.random.randint(1, 10, (4, 2))
print(A)
print(B)
product = np.dot(A, B)
print(product)
```

10. Generate an array of 10 random integers between 50 and 150 using Numpy. Find the minimum,
maximum, and the sum of the elements in the array.

```python
arr4 = np.random.randint(50, 151, 10)

min_val = arr4.min()
max_val = arr4.max()
sum_val = arr4.sum()

print("Array:", arr4)
print("Min:", min_val)
print("Max:", max_val)
print("Sum:", sum_val)
```
