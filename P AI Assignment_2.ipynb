Question 1:
Write a recursive function named is_palindrome(s) that checks whether the given string s is a
palindrome or not.
Requirements:
• The function should ignore spaces, punctuation, and case sensitivity.
• It should handle both even-length and odd-length strings (e.g., "madam", "abba", "A man a
plan a canal Panama").
• If the input string contains any numeric digits, the function should ignore them during
comparison but not remove them from the original string.
• The function should return True or False accordingly, and also print "Palindrome" or "Not
Palindrome" before returning.

```python
def is_palindrome(s):
    def clean_string(text):
        cleaned = ""
        for char in text:
            if char.isalpha():
                cleaned += char.lower()
        return cleaned

    def check(text):
        if len(text) <= 1:
            return True
        if text[0] != text[-1]:
            return False
        return check(text[1:-1])

    cleaned_str = clean_string(s)
    result = check(cleaned_str)

    if result:
        print("Palindrome")
    else:
        print("Not Palindrome")

    return result


print(is_palindrome("madam"))
print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("Hello 123"))
```

Question 2:
Write a recursive function sum_nested_list(lst) that calculates the sum of all numbers in a possibly
nested list, e.g.:
lst = [1, [2, [3, 4], 5], 6]

```python
def sum_nested_list(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested_list(item)
        else:
            total += item
    return total

print(sum_nested_list([1, [2, [3, 4], 5], 6]))
```

Question 3:
Create a text file named students.txt that stores the records of 10 students.
Each line in the file should contain the following information, separated by commas:
Name, Age, Grade
Example:

Ali, 20, 88
Write a program to:
• Read the file line by line.
• Use an if-else condition to categorize each student as “Above Average” (grade ≥ 80) or
“Below Average”.
• Store results in a dictionary and save them to a new file named result.txt.

```python
def process_student_records():
    students = [
        "Ali, 20, 88",
        "Sara, 19, 75",
        "Usman, 21, 92",
        "Ayesha, 20, 68",
        "Hassan, 22, 81",
        "Fatima, 19, 79",
        "Bilal, 23, 85",
        "Zara, 20, 73",
        "Ahmed, 21, 90",
        "Noor, 22, 77"
    ]

    with open("students.txt", "w") as f:
        for record in students:
            f.write(record + "\n")

    results = {}

    with open("students.txt", "r") as f:
        for line in f:
            name, age, grade = line.strip().split(",")
            grade = int(grade)
            if grade >= 80:
                results[name.strip()] = "Above Average"
            else:
                results[name.strip()] = "Below Average"

    with open("result.txt", "w") as f:
        for name, category in results.items():
            f.write(f"{name}: {category}\n")

    for name, category in results.items():
        print(f"{name}: {category}")

process_student_records()
```

Question 4:
Define a function process_numbers(nums) that:
• Take a list of 15 integer elements from user, list have duplicate values.
• Removes duplicates, sorts the list, and returns a tuple of (max, min, sum, average) using built-in
functions.
• Call this function in the main program and print results.

```python
def process_numbers(nums):
    nums = list(set(nums))
    nums.sort()
    maximum = max(nums)
    minimum = min(nums)
    total = sum(nums)
    average = total / len(nums)
    return (maximum, minimum, total, average)

nums = []
for i in range(15):
    n = int(input(f"Enter number {i+1}: "))
    nums.append(n)

result = process_numbers(nums)
print("Results (Max, Min, Sum, Average):", result)
```

Question 5:
Write a program that:
• Reads a text file named "data.txt".
• Counts how many times each unique word appears using a dictionary.
• Ignore punctuation and case sensitivity.
• Print the five most frequent words along with their counts.

```python
def count_words():
    with open("data.txt", "r") as f:
        text = f.read()

    text = text.lower()
    for ch in ",.!?;:'\"()-":
        text = text.replace(ch, "")

    words = text.split()
    word_count = {}

    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

    print("Top 5 most frequent words:")
    for word, count in sorted_words[:5]:
        print(f"{word}: {count}")

with open("data.txt", "w") as f:
    f.write("This is a sample text file.\n")
    f.write("It contains some sample text.\n")
    f.write("This text will be used to count the words.\n")
count_words()
```

Question 6:
Write a recursive function count_vowels(s) that:
• Counts and returns the number of vowels in a string s.
• Ignore spaces and convert all letters to lowercase before counting.
• Use an if-else condition to handle the base case.

```python
def count_vowels(s):
    s = s.replace(" ", "").lower()
    vowels = "aeiou"

    if s == "":
        return 0
    else:
        if s[0] in vowels:
            return 1 + count_vowels(s[1:])
        else:
            return count_vowels(s[1:])

text = input("Enter a string: ")
print("Number of vowels:", count_vowels(text))
```

Question 7:
Create a program that:
• Continuously asks the user to enter numbers until they type 'stop'.
• Uses a while loop to calculate and display the average of all even numbers entered.
• If no even number is entered, display an appropriate message.

```python
def average_even_numbers():
    numbers = []
    while True:
        value = input("Enter a number (or type 'stop' to finish): ")
        if value.lower() == "stop":
            break
        if value.isdigit() or (value.startswith('-') and value[1:].isdigit()):
            num = int(value)
            if num % 2 == 0:
                numbers.append(num)
        else:
            print("Invalid input. Please enter a number or 'stop'.")

    if numbers:
        avg = sum(numbers) / len(numbers)
        print("Average of even numbers:", avg)
    else:
        print("No even numbers were entered.")

average_even_numbers()
```

Question 8:
Given a list of strings, write a program to:
• Use a for loop to reverse each string only if its length is even.
• Store the modified strings in a tuple.
• Print both the original and final tuple.

```python
def reverse_even_length(strings):
    modified = []
    for s in strings:
        if len(s) % 2 == 0:
            modified.append(s[::-1])
        else:
            modified.append(s)
    return tuple(modified)

original = ("apple", "banana", "pear", "grapes", "kiwi", "melon")
result = reverse_even_length(original)

print("Original Tuple:", original)
print("Modified Tuple:", result)
```

Question 9:
Write a Python program that takes a dictionary of students and their marks as input (e.g., {'Ali': 85,
'Sara': 92, 'Bilal': 45}) and:
• Uses if-elif-else to categorize each student as ‘Pass with Distinction’, ‘Pass’, or ‘Fail’.
• Store the categorized results in a new dictionary.
• Finally, display the total count of students in each category.

```python
def categorize_students(students):
    categories = {}

    for name, marks in students.items():
        if marks >= 80:
            categories[name] = "Pass with Distinction"
        elif marks >= 50:
            categories[name] = "Pass"
        else:
            categories[name] = "Fail"

    distinction = sum(1 for c in categories.values() if c == "Pass with Distinction")
    passed = sum(1 for c in categories.values() if c == "Pass")
    failed = sum(1 for c in categories.values() if c == "Fail")

    print("Student Categories:")
    for name, category in categories.items():
        print(f"{name}: {category}")

    print("\nTotal Count:")
    print("Pass with Distinction:", distinction)
    print("Pass:", passed)
    print("Fail:", failed)

students = {'Ali': 85, 'Sara': 92, 'Bilal': 45, 'Hassan': 67, 'Zara': 78, 'Usman': 49}
categorize_students(students)
```
